## 2.2 Laplacian of Gaussian

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

# Import image
img_woman = cv2.imread('data/woman.png', cv2.IMREAD_GRAYSCALE)

# Denoise using Gaussian filter
denoised_woman = cv2.GaussianBlur(img_woman, (5,5), 0)

# Apply Laplacian filter
lapl_woman = cv2.Laplacian(denoised_woman, cv2.CV_64F, ksize=5)

# Compute threshold
max_val = np.max(lapl_woman)

_, img_thresh = cv2.threshold(lapl_woman, max_val*0.1, 255, cv2.THRESH_BINARY)
```

```python
# Display images
plt.figure(figsize=(20, 20))
plt.subplot(131), plt.imshow(img_woman, cmap="gray")
plt.title('Original'), plt.axis('off')
plt.subplot(132), plt.imshow(lapl_woman, cmap="gray")
plt.title('Laplacian of Gaussian'), plt.axis('off')
plt.subplot(133), plt.imshow(img_thresh, cmap="gray")
plt.title('Laplacian of Gaussian 10% Threshold'), plt.axis('off')
plt.show()
```
