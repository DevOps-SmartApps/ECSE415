## 2.2 Laplacian of Gaussian

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

# Import image
img_woman = cv2.imread('data/woman.png', cv2.IMREAD_GRAYSCALE)

# Denoise using Gaussian filter
denoised_woman = cv2.GaussianBlur(img_woman, (5,5), 0)

# Apply Laplacian filter
lapl_woman = cv2.Laplacian(denoised_woman, cv2.CV_32F, ksize=5)
```

```python
# Display images
plt.figure(figsize=(20, 20))
plt.subplot(131), plt.imshow(img_woman, cmap="gray")
plt.title('Original'), plt.axis('off')
plt.subplot(132), plt.imshow(denoised_woman, cmap="gray")
plt.title('Gaussian'), plt.axis('off')
plt.subplot(133), plt.imshow(lapl_woman, cmap="gray")
plt.title('Gaussian + Laplacian'), plt.axis('off')
plt.show()
```
